```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Write the code for simple imputing [[1, 2], [np.nan, 3], [7, 6]]

```python
data  = [[1,2],[np.nan, 3], [7, 6]]
df = pd.DataFrame(data)
df = df.fillna(df.mean())
print(df.head())
```

2.    Apply MinMaxScaler to [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]

```python
data2 = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
df2 = pd.DataFrame(data2)
df_scaled = (df2 - df2.min())/(df2.max() - df2.min())
print(df_scaled)
```

```python
iris_df = pd.read_csv("iris (6).csv")
print("First 5 rows\n")
print(iris_df.head())
print("Last 5 rows\n")
print(iris_df.tail())
```

7.    Calculate the mean of the feature ‘petallength’

```python
mean_petallength = iris_df['petallength'].mean()
print(mean_petallength)
```

```python
std_petallength = iris_df['petalwidth'].std()
print(std_petallength)
```

9.    Plot histograms of sepalwidth, petalwidth, sepallength, petallength and mention the axes

```python
fig, ax = plt.subplots(2,2)
fig.set_size_inches(10,10)

ax[0,0].hist(iris_df['sepalwidth'], bins = 20)
ax[0,0].set_xlabel('sepalwidth')
ax[0,0].set_ylabel('frequency')

ax[0,1].hist(iris_df['petalwidth'], bins = 20)
ax[0,1].set_xlabel('petalwidth')
ax[0,1].set_ylabel('frequency')

ax[1,0].hist(iris_df['sepallength'], bins = 20)
ax[1,0].set_xlabel('sepallength')
ax[1,0].set_ylabel('frequency')

ax[1,1].hist(iris_df['petallength'], bins = 20)
ax[1,1].set_xlabel('petallength')
ax[1,1].set_ylabel('frequency')
plt.show()
```

10. Print scatter plots (using seaborn library) with the following configurations for the iris data:

```python
fig, axs = plt.subplots(2,2)
fig.set_size_inches(10,10)

sns.scatterplot(x = 'petallength', y = 'sepalwidth', hue = 'class', data = iris_df, ax = axs[0,0])
sns.scatterplot(x = 'sepallength', y = 'sepalwidth', hue = 'class' , data = iris_df, ax = axs[0,1])
sns.scatterplot(x = 'sepallength', y = 'petalwidth', hue = 'class' , data = iris_df, ax = axs[1,0])
sns.scatterplot(x = 'petallength', y = 'petalwidth', hue = 'class' , data = iris_df, ax = axs[1,1])
```
